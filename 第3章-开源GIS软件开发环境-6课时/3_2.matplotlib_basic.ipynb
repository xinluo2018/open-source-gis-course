### matplotlib绘图基础
介绍画布figure, 子图axes, 子图坐标系Axis，子图标题title等。


```python
import matplotlib.pyplot as plt
import numpy as np
```

1. 画布画板示意

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('title', size=20)
ax.set_xlabel('xlabel', size=20)
ax.set_ylabel('ylabel', size=20)
plt.xticks(ticks=[0.2, 0.4, 0.6, 0.8], labels=['A', 'B', 'C', 'D'])
ax.tick_params(axis='x', labelsize=14)
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,4))
# fig, axes = plt.subplots(1, 2, figsize=(10,4))
```

2. 绘制线性图

```python
### 序列数据可视化
score = [89,94,78,82]
stud = ['zhang1', 'zhang2', 'zhang3', 'zhang4']
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].plot(score, marker='.', color='red', markersize=10)
# ax[0].plot(stud, score, marker='.', color='red', markersize=10)
```

```python
### 散点数据可视化
fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.scatter(x=stud, y=score, s=60, marker='*', color='green')
```

```python
### 柱状图
fig, ax = plt.subplots(1,1, figsize=(5,4))
ax.bar(x=stud, height=score, width=0.4, color='red')
```

```python
### 图像数据可视化
img = np.random.randn(100,100)
# img = np.random.randn(100,100, 3)
img.shape
fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.imshow(img, cmap='grey')
```
