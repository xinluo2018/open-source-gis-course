Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_axes([0,0,1,1])
ax.set_title('title', size=30)
ax.set_xlabel('xlabel', size=20)
ax.tick_params(axis='x', labelsize=24)
```

线性图

```python
data = [1,3,4,6]
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(data, color='red', marker='*', markersize=20)

```

散点图

```python
data_x = [1,2,3,4]
data_y = [3,2,1,4]
data_z = [200,70,40,90]

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.scatter(x=data_x, y=data_y, s=data_z, marker='*', color='green')
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,4))
data_scores = [87, 92, 79, 82]
data_students = ['zhang1', 'zhang2', 'zhang3', 'zhang4']
ax.bar(data_students, data_scores, color='green', width=0.4)
```
